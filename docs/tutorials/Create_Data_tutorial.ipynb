# Creating data with pyRANN

## 2) Elastic Strain
First, let's load an FCC Al structure from our POSCAR file, "Al.POSCAR".

```python
import pyrann as rann
import numpy as np

fcc = rann.load('Al.POSCAR')
print(fcc)
```

Now we can apply an elastic strain on our system.
We will generate 100 unique structures for each of the following strain modes in Voigt notation:
| $\varepsilon_1$ | $\varepsilon_2$ | $\varepsilon_3$ | $\varepsilon_4$ | $\varepsilon_5$ | $\varepsilon_6$ |
|---|---|---|---|---|---|
| $\varepsilon$ | $\varepsilon$ | $\varepsilon$ | 0 | 0 | 0 |
| $\varepsilon$ | 0 | 0 | 0 | 0 | 0 |
| 0 | $\varepsilon$ | 0 | 0 | 0 | 0 |
| 0 | 0 | $\varepsilon$ | 0 | 0 | 0 |
| 0 | 0 | 0 | $\varepsilon$ | 0 | 0 |
| 0 | 0 | 0 | 0 | $\varepsilon$ | 0 |
| 0 | 0 | 0 | 0 | 0 | $\varepsilon$ |
| 0 | 0 | 0 | $\varepsilon$ | $\varepsilon$ | $\varepsilon$ |
| $\varepsilon$ | $\varepsilon$ | 0 | 0 | 0 | 0 |
| $\varepsilon$ | $-\varepsilon$ | 0 | 0 | 0 | 0 |
| $\varepsilon$ | $\varepsilon$ | $\varepsilon$ | $\varepsilon$ | $\varepsilon$ | $\varepsilon$ |

We will apply a maximum strain of 5%. This means, for each strain mode,
$\varepsilon$ will vary from -5% to +5% strain in equal increments across the 100 unique structures.

```python
strain = fcc.strain(strain_points=100, maximum_strain = 0.05)
```

"strain" is now an instance of the series class. A series class object contains a list of system class objects.
We can export the strain series instance to a file.

```python
strain.export('strain.dump')
```

If we wanted individual POSCAR files for each unique structure, we can do the following:

```python
strain.export('strain.POSCAR', directory='POSCARS')
```

## 3) Create Supercell
For other types of data, we may want more than just a primitive cell. Let's create a supercell.

```python
new_box = fcc.box*np.array([4, 4, 4])
fcc.createsupercell(new_box=new_box)
print(fcc)
```

The createsupercell method modifies the attributes of the system instance passed to it. 
This means you will need to load in a new instance of "Al.POSCAR" if you need the primitive cell again.

## 4) Thermal Perturbations
Now that we have a supercell, let's generate data to mimic thermal perturbations.

We will create 100 "unique" structures with a maximum perturbation of 5% of Al lattice constant ($4.0495 \mathring{A} *0.05=0.202475 \mathring{A}$).
We will also randomly strain the box by 5% and export to a LAMMPS dump file.

```python
thermal = fcc.thermal(temp=0.202475, strain=0.05, nsims=100)
thermal.export('thermal.dump')
```

## 5) Vacancies
Now let's add a vacancy to our supercell. We do this by randomly deleting one atom.

```python
fcc.vacancy(num_neighbors=1)
print(fcc)
```

Notice that the vacancy method modifies the attributes of the fcc system instance. This means that any operation
done on "fcc" will now be done on an FCC system containing a vacancy. Let's induce thermal perturbations on this 
system.

```python
thermal_vacancy = fcc.thermal(temp=0.202475, strain=0.05, nsims=100)
thermal_vacancy.export('thermal_vacancy.dump')
```

We can increase the size of the vacancy by increasing the num_neighbors argument.

```python
fcc.vacancy(num_neighbors=3)
print(fcc)
```

Notice how "fcc" now only has 60 atoms. This is because we ran the vacancy method on it twice. The first time, "fcc" went from 64 atoms to 63 atoms because we set num_neighbors=1. The second time we ran the vacancy method, we set num_neighbors=3, so "fcc" went from 63 atoms to 60 atoms.
